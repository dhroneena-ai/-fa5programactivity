```python
def calculate_total_price(price_per_item, quantity):
    total_price = price_per_item * quantity
    return total_price

price = int(input("Enter the price of one item: "))
num_items = int(input("Enter the number of items purchased: "))
total = calculate_total_price(price, num_items)

print(f"The total price for {num_items} items at ₱{price} each is: ₱{total}")
```
